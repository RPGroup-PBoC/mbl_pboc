```python
# import some modules
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline

# for nice plotting
import seaborn as sns
sns.set()
```

```python
# biological parameters 
k = 1/20 # per min
N0 = 1

# simulation parameters
dt = 0.1 # min 
total_time = 120 # min
```

```python
num_steps = int(total_time/dt)

# initilize an array to store values of N over time
N_t = np.zeros(num_steps)
N_t[0] = N0
```

```python
# loop through time steps 
for t in range(1, num_steps):
   
    # first calculate dN
    dN = N_t[t-1] * k * dt
    
    # update current entry
    N_t[t] = N_t[t-1] + dN
```

```python
# x axis values
times = np.arange(num_steps)*dt # mins

# plot!
plt.plot(times, N_t, label="simulation")
plt.xlabel("time (mins)")
plt.ylabel("N")

# compute the "true" curve
solution = N0 * np.exp(k*times)

# plot the "true" curve on top
plt.plot(times, solution, label="solution")
plt.legend()
```
